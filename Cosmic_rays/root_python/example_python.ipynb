```python
import ROOT
from array import array
from ROOT import TFile, TCanvas, TTree, TH1D, TH2D, TRandom
```

```python
#creating a root file with radom values and two branches
file = TFile("my_file.root", "recreate")
tree = TTree("my_tree", "my_tree")

rnd = TRandom()

maxn = 10
x = array('f', [ 0. ])
y = array('f', [ 0. ])
tree.Branch('x', x, 'x/F')
tree.Branch('y', y, 'y/F')

for i in range(20000):
    x[0] = rnd.BreitWigner()
    y[0] = rnd.BreitWigner()*x[0]
    tree.Fill()

file.Write()
file.Close()
    
```

```python
#opening the root file
new_file = TFile("my_file.root")
new_tree = new_file.Get("my_tree")

#declaring the histograms
histo1D = TH1D("histo1D", "1 dim histogram", 1000, -1, 1)
histo2D = TH2D("histo2D", "2 dim histogram", 1000, -1, 1, 1000, -1, 1)

#Filling the histograms
for entry in new_tree:
    histo2D.Fill(entry.x, entry.y)
    histo1D.Fill(entry.x)
```

```python
#enable interactive visualization
ROOT.enableJSVis()

#plotting one histo
cv1D = TCanvas("my_cv")
histo1D.Draw()
cv1D.Draw()

#plotting the second histo
cv2D = TCanvas("my_cv2")
histo2D.Draw("colz")
cv2D.Draw()
```
